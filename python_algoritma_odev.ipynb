```python
# Verilen bir tamsayı listesindeki en büyük sayıyı bulan bir fonksiyon yazın
liste = [3, 5, 2, 8, 1, 4]

def en_buyuk_sayi(liste):
    return max(liste)

print(en_buyuk_sayi(liste))
```

```python
# Verilen bir string'i ters çeviren bir fonksiyon yazın.
def ters_cevir(s):
    return s[::-1]

print(ters_cevir("Merhaba"))
```

```python
# Negatif olmayan bir tamsayının faktöriyelini hesaplayan bir fonksiyon yazın.
def faktoriyel(n):
    if n == 0:
        return 1 
    sonuc = 1
    for i in range(1,n+1):
        sonuc *= i
    return sonuc

print(faktoriyel(5))
```

```python
#Verilen bir tamsayı listesindeki yalnızca çift sayıları içeren yeni bir liste döndüren bir fonksiyon yazın.
def ciftleri_filtrele(liste):
    return [sayi for sayi in liste if sayi % 2 == 0]

print(ciftleri_filtrele([1, 2, 3, 4, 5, 6]))
```

```python
#Verilen iki listeyi birleştiren ve tekrar eden elemanları kaldırarak tekil bir liste oluşturan bir fonksiyon yazın.
def birlestir_ve_tekillestir(liste1,liste2):
    birlesik_liste = liste1+liste2
    return list(set(birlesik_liste))

print(birlestir_ve_tekillestir([1,2,3],[3,4,5]))
```

```python
#Verilen bir string'in palindrom olup olmadığını kontrol eden bir fonksiyon yazın (tersten okunuşu aynı olan kelime/cümle).
def is_palindrome(s):
    temiz_s = ''.join(filter(str.isalnum,s)).lower()
    return temiz_s == temiz_s[::-1]

print(is_palindrome("A man, a plan, a canal: Panama"))
```

```python
#Verilen iki pozitif tamsayının en büyük ortak bölenini (EBOB) bulan bir fonksiyon yazın.
def ebob(a,b):
    while b:
        a,b = b,a%b
    return a

print(ebob(48,18))
```

```python
#Verilen iki string'in anagram olup olmadığını kontrol eden bir fonksiyon yazın (aynı harfleri farklı sıralamada içeren kelimeler).
def is_anagram(s1,s2):
    temiz_s1 = sorted(''.join(filter(str.isalnum,s1)).lower())
    temiz_s2 = sorted(''.join(filter(str.isalnum,s2)).lower())
    return temiz_s1 == temiz_s2

print(is_anagram("listen","silent"))
```

```python
#1'den N'e kadar olan sayılardan birinin eksik olduğu bir liste veriliyor. Eksik olan sayıyı bulan bir fonksiyon yazın.
def find_missing_number(nums):
    n = len(nums)
    beklenen_toplam = n * (n+1) //2
    gercek_toplam = sum(nums)
    return beklenen_toplam - gercek_toplam

print(find_missing_number([0,1,2,4,5]))  
```

```python
#Verilen bir listeyi, belirtilen k adımı kadar sağa döndüren bir fonksiyon yazın.
def rotate_array(nums,k):
    n = len(nums)
    k = k % n
    if k == 0 : 
        return nums
    return nums[-k:] + nums[:-k]

print(rotate_array([1,2,3,4,5],2))
```

```python
#Verilen iki sıralı tamsayı listesinin birleştirilmiş halinin medyanını bulan bir fonksiyon yazın. Algoritmanın zaman karmaşıklığı O(log(m+n)) olmalıdır; burada m ve n listelerin uzunluklarıdır.
def findMedianSortedArrays(nums1,nums2):
    if len(nums1) > len(nums2):
        nums1,nums2 = nums2,nums1
    m,n = len(nums1),len(nums2)
    left,right = 0,m
    while left <= right:
        partition1 = (left + right) // 2
        partition2 = (m+n+1) // 2 - partition1
        
        max_left1 = float("-inf") if partition1 == 0 else nums1[partition1 - 1]
        max_left2 = float('-inf') if partition2 == 0 else nums2[partition2 - 1]

        min_right1 = float('inf') if partition1 == m else nums1[partition1]
        min_right2 = float('inf') if partition2 == n else nums2[partition2]

        if max_left1 <= min_right2 and max_left2 <= min_right1:
            # Eğer toplam eleman sayısı çiftse, ortadaki iki sayının ortalamasını döndür
            if (m + n) % 2 == 0:
                return (max(max_left1, max_left2) + min(min_right1, min_right2)) / 2
            # Eğer toplam eleman sayısı tekse, sol tarafın en büyüğü medyandır
            else:
                return max(max_left1, max_left2)
        
        # Sol tarafta fazla eleman varsa, partition1'i sola kaydır
        elif max_left1 > min_right2:
            right = partition1 - 1
        # Sol tarafta az eleman varsa, partition1'i sağa kaydır
        else:
            left = partition1 + 1


# Test örneği
print(findMedianSortedArrays([1, 3], [2]))  # Beklenen çıktı: 2.0
```

```python
#Verilen bir string içindeki en uzun palindromik alt string'i bulan bir fonksiyon yazın.
def longestPalindrome(s):
    if not s:
        return ""
    start = 0       # En uzun palindromun başlangıç indeksini tutar
    max_len = 1     # Şu ana kadar bulunan en uzun palindromun uzunluğu

    # Merkezden genişleyerek palindrom bulan yardımcı fonksiyon
    def expand_around_center(left, right):
        # Sol ve sağdan dışa doğru genişler, karakterler eşit olduğu sürece devam eder
        while left >= 0 and right < len(s) and s[left] == s[right]:
            left -= 1   # Solu bir adım geri kaydır
            right += 1  # Sağı bir adım ileri kaydır
        # Döngüden çıkınca bir adım fazla genişlemiş oluruz,
        # bu yüzden palindromun gerçek uzunluğu (right - left - 1)
        return right - left - 1

    # String'in her karakterini merkez kabul ederek genişleme yap
    for i in range(len(s)):
        # Tek merkezli palindromlar için (örneğin "aba" -> merkez: 'b')
        len1 =expand_around_center(i,i)
        # Çift merkezli palindromlar için (örneğin "abba" -> merkez: 'bb')
        len2 = expand_around_center(i,i+1)

        # O andaki en uzun palindromun uzunluğunu al
        current_max = max(len1,len2)
        
        # Yeni bulunan palindrom, önceki en uzundan uzunsa güncelle
        if current_max > max_len:
            max_len = current_max
            # Yeni başlangıç indeksini hesapla
            # (current_max - 1) // 2 ifadesi, merkezden sola kaymayı ayarlar
            start = i - (current_max - 1) // 2

    # En uzun palindromu orijinal string içinde dilimle
    return s[start:start + max_len]

print(longestPalindrome("babad"))  # Çıktı: "bab" veya "aba"



```

```python
from collections import deque

def ladderLength(beginWord, endWord, wordList):
    if endWord not in wordList:
        return 0
    
    wordSet = set(wordList)
    queue = deque([(beginWord, 1)])  # (kelime, adım_sayısı)
    visited = set([beginWord])
    
    while queue:
        current_word, steps = queue.popleft()
        
        # Hedef kelimeye ulaştık
        if current_word == endWord:
            return steps
        
        # Mevcut kelimeden bir harf değiştirerek yeni kelimeler üret
        for i in range(len(current_word)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                new_word = current_word[:i] + c + current_word[i+1:]
                
                # Yeni kelime geçerli ve daha önce ziyaret edilmemişse
                if new_word in wordSet and new_word not in visited:
                    visited.add(new_word)
                    queue.append((new_word, steps + 1))
    
    return 0  # Dönüşüm mümkün değil

# Test
print(ladderLength("hit", "cog", ["hot","dot","dog","lot","log","cog"]))
# Çıktı: 5
```

```python
#Verilen bir tamsayı dizisindeki bitişik alt dizilerden en büyük toplama sahip olanını bulun ve bu toplamı döndürün.
def maxSubArray(nums):
    # İlk elemanı başlangıç değeri olarak alıyoruz
    # max_current: o ana kadar olan alt dizinin maksimum toplamı
    # max_global: genel (tüm dizideki) maksimum toplam
    max_current = max_global = nums[0]

    # Dizinin 2. elemanından itibaren dolaş
    for i in range(1, len(nums)):
        # Mevcut elemanı (nums[i]) önceki toplamla birleştirip devam etmek mi,
        # yoksa yeni bir alt dizi başlatmak mı daha iyi diye karar veriyoruz
        # Eğer önceki toplam negatifse, yeni alt dizi başlatmak daha mantıklı olur
        max_current = max(nums[i], max_current + nums[i])
        # Şimdiye kadarki en büyük toplamı (global maksimum) güncelliyoruz
        max_global = max(max_global, max_current)
    
    # Döngü bittiğinde, max_global dizideki en büyük alt dizi toplamını tutar
    return max_global

print(maxSubArray([-2, 1, -3, 4, -1, 2, 1, -5, 4]))  # Çıktı: 6
```

```python
#Verilen bir aralık listesinde, çakışan aralıkları birleştirin ve yeni bir aralık listesi döndürün.
def merge(intervals):
    # Eğer liste boşsa, direkt boş bir liste döndür
    if not intervals:
        return []

    # Aralıkları (intervals) başlangıç değerine göre sırala
    # Örneğin [[2,6],[1,3]] → [[1,3],[2,6]] olur

    intervals.sort(key=lambda x: x[0])

    # İlk aralığı birleştirilmiş listeye ekle
    merged = [intervals[0]]
    
    # Kalan aralıkları sırayla dolaş
    for current in intervals[1:]:
        # merged[-1] → şu ana kadar birleştirilen son aralık
        last_merged = merged[-1]
        
        # Eğer mevcut aralığın başlangıcı (current[0]),
        # son birleştirilmiş aralığın bitişinden küçük veya eşitse
        # bu iki aralık çakışıyor demektir.
        if current[0] <= last_merged[1]:
            # Çakıştığı için, bitiş değerini güncelle
            # hangisinin bitişi büyükse onu al (daha geniş aralığı kapsamak için)
            last_merged[1] = max(last_merged[1], current[1])
        else:
            # Çakışmıyorsa, yeni bir aralık olarak ekle
            merged.append(current)
    
    # Tüm aralıklar işlenince, birleşmiş aralıkları döndür
    return merged


print(merge([[1,3],[2,6],[8,10],[15,18]]))  # Çıktı: [[1, 6], [8, 10], [15, 18]]

```
